```python
import spacy
TA = spacy.load("en_core_web_sm")
text = "Barack Obama was born in Hawaii and worked at Google."
doc = TA(text)
for ent in doc.ents:
    print(ent.text, "-" ,ent.label_)
```

```python
import spacy
from docx import Document

nlp = spacy.load("en_core_web_sm")

def get_docx_text(path):
    doc = Document(path)
    return "\n".join([para.text for para in doc.paragraphs])

full_text = get_docx_text("War.docx")

doc = nlp(full_text)
for ent in doc.ents:
    print(f"{ent.text:<20} | {ent.label_}")
```
