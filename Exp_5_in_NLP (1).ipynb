```python
import math
from collections import Counter

doc1 = "B Tech"
doc2 = "B Tech"

words1 = doc1.lower().split()
words2 = doc2.lower().split()

set1 = set(words1)
set2 = set(words2)

intersection = set1.intersection(set2)
union = set1.union(set2)

jaccard_similarity = len(intersection) / len(union)

print(jaccard_similarity)


counter1 = Counter(words1)
counter2 = Counter(words2)

all_words = list(set(words1 + words2))

vec1 = [counter1[word] for word in all_words]
vec2 = [counter2[word] for word in all_words]

dot_product = sum(x * y for x, y in zip(vec1, vec2))

magnitude1 = math.sqrt(sum(x ** 2 for x in vec1))
magnitude2 = math.sqrt(sum(x ** 2 for x in vec2))

cosine_similarity = dot_product / (magnitude1 * magnitude2)
print("Cosine Similarity:", cosine_similarity)
```
