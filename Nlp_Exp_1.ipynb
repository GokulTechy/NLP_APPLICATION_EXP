```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

1.Tokenization = Word tokenization is the most commonly used method where text is divided into individual words


```python
text =" It is obvious that people want to get the best services for the best price. You should also keep in mind that bigger price does mean better quality!@. "
print(f"ORIGINAL TEXT:\n{text}\n")
tokens = word_tokenize(text)
print("1. TOKENIZATION (Splitting  words):")
print(tokens, "\n")
```

2.Removing The Stopwords



```python
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words and word.isalpha()]
print("2. STOPWORDS REMOVAL (Filtering words):")
print(filtered_tokens, "\n")
```

3. Stemming = Root Word

```python
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_tokens]
print("3. STEMMING :")
print(stemmed_words, "\n")
```

4.Lemmatization = Convert Into Dict Words

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print("4. LEMMATIZATION (Converting to real dictionary base words):")
print(lemmatized_words, "\n")
```
