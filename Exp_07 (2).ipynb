```python
!pip install transformers torch sentencepiece

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Load Model
tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn")

# Input Text
text = """
Tamil is one of the oldest living languages in the world and has a rich cultural and literary heritage.
It is primarily spoken in the Indian state of Tamil Nadu and in countries such as Sri Lanka, Singapore,
and Malaysia. Tamil belongs to the Dravidian language family and has a recorded history of more than
two thousand years.

Tamil literature is famous for its classical works, especially Sangam literature, which reflects the
culture, traditions, and lifestyle of ancient Tamil society. The language has contributed greatly to
poetry, philosophy, music, and art. Many ancient temples in Tamil Nadu contain inscriptions written in
Tamil that provide valuable historical information.

Tamil is also widely used in modern technology, cinema, education, and media. The Tamil film industry,
popularly known as Kollywood, is one of the largest film industries in India. Tamil songs, movies, and
literature are appreciated by people around the world.

The Government of India declared Tamil as a Classical Language because of its ancient origin and
independent literary tradition. Today, millions of people across the globe continue to speak and preserve
the Tamil language and its unique cultural identity.
"""

# Prepare input for the model
inputs = tokenizer([text], max_length=1024, return_tensors="pt", truncation=True)

# Generate Summary
summary_ids = model.generate(inputs["input_ids"], num_beams=4, max_length=100, min_length=30, early_stopping=True)
summary_text = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# Output
print("\n===== GENERATED SUMMARY =====\n")
print(summary_text)
```
